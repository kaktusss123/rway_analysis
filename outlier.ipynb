# Поиск отклонений

## Необходимые импорты

```python
import sqlalchemy
import pandas as pd
engine = sqlalchemy.create_engine('postgresql://rway:rway@10.199.13.111/rway_analysis')
```

## Считывание с базы

```python
test = pd.read_sql('SELECT * FROM fill_rate', con=engine, index_col='index')
test
```

## Поиск отклонений от среднего значения по старым данным

```python
def deviation(df):
    max_date = max(df['ДатаЗадания'])
    last = df.loc[df.loc[:, 'ДатаЗадания'] == max_date]
    df_ = pd.DataFrame(index=last.index, columns=last.columns)
    last.drop(['ДатаЗадания', 'КодЗадания', 'КодЗадачи'], axis=1, inplace=True)
    
    for ind in last.index:
        for col in last.columns:
            try:
                df_.loc[ind, col] = last.loc[ind, col] - df.loc[ind, col].mean()
            except Exception as e:
                print('Exception: {}'.format(e))
                df_.loc[ind, col] = 0
    return df_
```

```python
dev = deviation(test)
dev
```

На всякий случай удалим дубликаты строк из базы

```python
dev = dev.drop_duplicates()
```

## Поиск характеристик, значения которых сильно отклонились от среднего
В переменную `threshold` записывается пороговое значение отклонения в процентах. Результат поиска записывается в файл `outliers.csv`.

```python
threshold = 5
```

```python
ds = dev.stack()
outliers = ds[abs(ds) > threshold].to_frame()
outliers.columns = ['Отклонение']
outliers.to_csv('outliers.csv', encoding='cp1251', sep=';')
outliers
```
